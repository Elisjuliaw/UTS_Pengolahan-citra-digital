```python
# Install OpenCV dan Matplotlib
!pip install opencv-python-headless matplotlib

# Import pustaka yang dibutuhkan
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Memuat gambar yang sudah diunggah
original_image = cv2.imread(image_path)

# Memeriksa apakah gambar berhasil dimuat
if original_image is None:
    print("Gambar tidak ditemukan atau gagal dimuat.")
else:
    print("Gambar berhasil dimuat.")

    # Konversi gambar ke grayscale
    gray_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)

    # Apply histogram equalization
    equalized_image = cv2.equalizeHist(gray_image)

    # Ubah kembali menjadi BGR untuk tampilan side-by-side
    equalized_bgr = cv2.cvtColor(equalized_image, cv2.COLOR_GRAY2BGR)

    # Fungsi untuk menampilkan gambar
    def show_images(original, equalized):
        plt.figure(figsize=(10, 5))
        plt.subplot(1, 2, 1)
        plt.imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB))
        plt.title("Original Image")

        plt.subplot(1, 2, 2)
        plt.imshow(cv2.cvtColor(equalized, cv2.COLOR_BGR2RGB))
        plt.title("Histogram Equalized Image")
        plt.show()

    # Tampilkan gambar asli dan hasil equalization
    show_images(original_image, equalized_bgr)

    # Menampilkan histogram asli dan hasil equalization
    def plot_histogram(image, title):
        plt.hist(image.ravel(), 256, [0, 256])
        plt.title(title)
        plt.xlabel("Intensity Value")
        plt.ylabel("Pixel Count")
        plt.show()

    plot_histogram(gray_image, "Histogram Original Image")
    plot_histogram(equalized_image, "Histogram Equalized Image")
```
